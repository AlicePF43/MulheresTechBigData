```python
import pandas as pd

dados = pd.read_csv('Top 100 Healthiest Food in the World.csv')

dados_copia = dados.copy()

display(dados.head())

display(dados_copia.head())

estatistica = dados.describe()

display(estatistica)
```

alimentos que têm mais de 20g de proteína e menos de 100 calorias por porção.

```python
# Selecionando alimentos com mais de 25g de proteína e menos de 100 calorias
high_protein_low_calories = dados.query('`Protein (g)` > 20 and Calories < 100')
display(high_protein_low_calories)
```

Alimentos que têm um Antioxidant Score maior que 5000 e mais de 20mg de Vitamina C

```python
# Selecionando alimentos com Antioxidant Score maior que 5000 e mais de 20mg de Vitamina C
high_antioxidants_vitamin_c = dados.query('`Antioxidant Score` > 5000 and `Vitamin C (mg)` > 20')
display(high_antioxidants_vitamin_c)
```

Alimentos que têm mais de 3g de fibra e menos de 50 calorias por porção.

```python
# Selecionando alimentos com mais de 3g de fibra e menos de 50 calorias
high_fiber_low_calories = dados.query('`Fiber (g)` > 3 and Calories < 50')
display(high_fiber_low_calories)
```
